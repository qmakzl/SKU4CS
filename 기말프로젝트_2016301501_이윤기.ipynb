```python
import pandas as pd
import numpy as np
from glob import glob
import matplotlib.pyplot as plt 
from matplotlib import font_manager, rc 
%matplotlib inline

font_path='C:/Windows/Fonts/malgun.ttf' 
font_name= font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font_name)
```

```python
def modify_data(df):
    target_cols=df.iloc[15].values
    change_cols=dict( [(org,new) for org, new in  zip(df.columns, target_cols)])
    df.rename( columns=change_cols, inplace=True)
    df= df[16:]
    df.reset_index(drop=True, inplace=True)
    return df
```

```python
op_df=pd.read_excel('./오피스텔_실거래가_20191215.xlsx', encoding='euc-kr')
op_df.head(20)
```

```python
result_op_df=modify_data(op_df)
result_op_df.head(5)
```

```python
ap_df=pd.read_excel('./아파트_실거래가_20191215.xlsx', encoding='euc-kr')
ap_df.head(20)
```

```python
result_ap_df=modify_data(ap_df)
result_ap_df.head(5)
```

```python
sum_df = pd.concat([result_op_df, result_ap_df], keys=['result_op_df','result_ap_df']) #2개를 합침
sum_df.head(5)
```

```python
del sum_df["번지"]
del sum_df["층"]
del sum_df["부번"]
del sum_df["본번"]
del sum_df["단지명"]
```
